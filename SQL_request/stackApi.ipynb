```python
# Installation de la bibliothèque StackAPI si nécessaire
from stackapi import StackAPI
import pandas as pd

# Initialisation de l'objet StackAPI pour Stack Overflow
SITE = StackAPI('stackoverflow')

# Définition des paramètres de la requête
params = {
    'pagesize': 50,
    'fromdate': '2020-01-01',  # Définir la date de début souhaitée
    'todate': '2023-12-31',    # Définir la date de fin souhaitée
    'order': 'desc',
    'sort': 'votes',
    'tagged': 'python',
    'min': 51,                 # Score > 50
    'site': 'stackoverflow'
}

# Récupération des données
questions = SITE.fetch('questions', **params)

# Extraction des données principales dans un DataFrame
data = []
for item in questions['items'][:50]:
    data.append({
        'date': item['creation_date'],
        'title': item['title'],
        'tags': item['tags'],
        'score': item['score']
    })

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'], unit='s')
display(df)

# Affichage du DataFrame
#import ace_tools as tools; tools.display_dataframe_to_user(name="Stack Overflow Questions", dataframe=df)
```
